```python
def clean_df(df):    
    copied_df["day"] = copied_df["date"].dt.day_name()
    copied_df["month"] = copied_df["date"].dt.month_name()
    copied_df["year"] = copied_df["date"].dt.year
    return df


def access(df):
    conditions = [
        (copied_df["day"] == "Monday"),
        (copied_df["day"] == "Tuesday"),
        (copied_df["day"] == "Wednesday"),
        (copied_df["day"] == "Thursday"),
        (copied_df["day"] == "Friday"),
        (copied_df["day"] == "Saturday"),
        (copied_df["day"] == "Sunday")
        ]

    # create a list of the values we want to assign for each condition
    values = ['Week day', 'Week day', 'Week day', 'Week day', 'Week day', 'Weekend', 'Weekend']

    # create a new column and use np.select to assign values to it using our lists as arguments
    copied_df['week_Class'] = np.select(conditions, values)
    
    return df

copied_df["participant_gender"] = copied_df["participant_gender"].fillna("0::Unknown")

def clean_participant_gender(row) :
    gender_row_values = []
    gender_row = str(row).split("||")
    for x in gender_row :
        gender_row_value = str(x).split("::")
        if len(gender_row_value) > 1 :
            gender_row_values.append(gender_row_value[1])
            
    return gender_row_values

participant_genders = copied_df.participant_gender.apply(clean_participant_gender)
copied_df["participant_gender_total"] = participant_genders.apply(lambda x: len(x))
copied_df["participant_gender_male"] = participant_genders.apply(lambda x: x.count("Male"))
copied_df["participant_gender_female"] = participant_genders.apply(lambda x: x.count("Female"))
copied_df["participant_gender_unknown"] = participant_genders.apply(lambda x: x.count("Unknown"))
del(participant_genders)


def positioning(df):
  columns= ['incident_id', 'date', 'day', 'month', 'year', 'week_Class', 'state', 'city_or_county', 'address', 'n_killed',
       'n_injured', 'incident_url', 'source_url',
       'incident_url_fields_missing', 'congressional_district', 'gun_stolen',
       'gun_type', 'incident_characteristics', 'latitude',
       'location_description', 'longitude', 'n_guns_involved', 'notes',
       'participant_age', 'participant_age_group', 'participant_gender', 'participant_gender_total',
       'participant_gender_male', 'participant_gender_female',
       'participant_gender_unknown',
       'participant_name', 'participant_relationship', 'participant_status',
       'participant_type', 'sources', 'state_house_district',
       'state_senate_district']
  pos_df = copied_df.reindex(columns=columns)
  return pos_df

copied_df = clean_df(copied_df)
copied_df = access(copied_df)
copied_df = positioning(copied_df)
copied_df.isna().sum()
copied_df = copied_df.replace('', np.nan)                   # to get rid of empty values
copied_df.isna().sum() 

copied_df.head(2)
# copied_df.to_csv("/content/drive.csv")
# copied_df.head(2)
                   
```
